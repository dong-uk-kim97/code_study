# 최초 시도

```python
T = list(map(int,input().split()))
i=0
for i in T:
  if T[i]>T[i+1]:
    print('Yes')
  else:
    print('No')
  i+=2
```

for 반복문에 range 설정을 안해줘서 오류가 났다. range 설정을 하고 다시 시도해본다.

```python
T = list(map(int,input().split()))
i=0
for i in range(len(T)):
  if T[i]>T[i+1]:
    print('Yes')
  else:
    print('No')
  i+=2
```

T를 리스트로 만들었는데 이게 잘못된 선택인 것 같다. 이것을 그냥 반복문 안에서 받도록 해야겠다.


```python
i=0
for i in T:
  T = map(int,input().split())
  if i>i+1:
    print('Yes')
  else:
    print('No')
  i+=2
```

반복문 안에 넣었던 것이 잘못인 것 같다. 반복문 안에 넣으면 하나 하나 입력을 해줘야 하는데 문제는 한번에 입력한다고 했으니 밖에서 차례대로 빼서 쓰는 방식으로 해야될 것 같다. 다시 반복문 밖으로 빼내고 map으로 시도해본다.

```python
T =list(map(int,input().split('\n')))

for i in range(1,len(T)):
  if T[i-1]>T[i]:
    print('Yes')
  else:
    print('No')
  i+=3
```

for 반복문을 사용하니 이상해지는 것 같다. while문으로 시도해본다.

# 정답

```python
while True:
  a,b=map(int,input().split())
  if a==0 and b==0:
    break
  elif a>b:
    print('Yes')
  else:
    print('No')
```

for 반복문에 집착하다 보니 뭔가 문제를 어렵게 푼 것 같다. 다음에는 다른 방식도 고려해가며 시도해봐야겠다.
