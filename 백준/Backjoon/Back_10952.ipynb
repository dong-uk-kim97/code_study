```python
# 백준 10952번 문제
while True:
  a,b=map(int,input().split())
  if (a and b) !=0:
    print(a+b)
  else: break
```

while True문 안으로 a와 b를 집어 넣어야 할 줄은 몰랐다.
while True로 무한 루프를 만들고 a와b 모두 무한 루프 안으로 집어 넣어야 코드가 실행이 가능하다는 것을 알게 됐다.
if 조건문을 통해서 a와 b가 아니면 더하고 그렇지 않은 경우 무한 루프를 중단하는 while문을 만들면서 조건을 충족시켰다.
코딩 테스트를 위한 준비의 첫 단계이다. 꾸준히 문제를 풀어서 올려서 점차 코딩 실력이 쌓였으면 하는 바램이다.
