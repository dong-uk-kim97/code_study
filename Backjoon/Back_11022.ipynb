# 오답

```python
T = int(input())
count = 0
while True:
    a,b = map(int, input().split())
    if T>=count:
        print('Case #', T, ':', a, '+', b, '=', a+b)
    count +=1
    else:
        break
```

우선 여기서는 컴파일 에러가 떴다. 그리고 오류 문장을 보니 문법에서 에러가 발생했다는 것을 알 수 있었다. 그 이유로는 T가 아닌 count를 써야했고, count+=1 부분은 if의 코드블록에 포함되어야 하는데 그렇지 못했다. 그래서 고쳐서 다시 제출해보았다.

```python
import sys
T = int(input())
count = 0
while True:
    a,b = map(int, input(sys.stdlin.readline().rstrip()).split())
    if T>=count:
        print('Case','#',count,':', a, '+', b, '=', a+b)
        count +=1
    else:
        break
```

코드 블록을 제출하니 이제 런타임에러가 떴다. 그래서 빠르게 입력 받는 sys.stdin.readline().rstrip()을 써서 제출을 했는데 이래도 런타임 에러가 떴다. 그래서 메모리를 많이 쓴다는 것을 깨닫고 while문을 쓰지 않기로 했다.

```python
T=int(input())
A, B = map(int, input().split())
for i in range(1, T+1):
    print('Case #',i,':',A,'+',B,'=',A+B)
```

for문으로 바꿨는데 A,B가 for문 안에서 반복시켜야 하는데 엉뚱하게 해버렸다. 그래서 고쳐서 다시 냈다.

```python
T=int(input())
for i in range(1, T+1):
    A, B = map(int, input().split())
    print('Case #',i,':',A,'+',B,'=',A+B)
```

A,B를 다시 넣고 제출해봤다. 그런데도 틀렸다고 나와서 i+=1을 안해서 그런가라는 의심을 했다(원래는 하면 안된다는 것을 알고 있었다). 그래서 넣고 진행해봤다.

```python
T=int(input())
for i in range(1, T+1):
    A, B = map(int, input().split())
    print('Case #',i,':',A,'+',B,'=',A+B)
    i+=1
```

```python
T=int(input())
for i in range(1, T+1):
    A, B = map(int, input().split())
    print('Case #',i,':' ,A ,'+' ,B ,'=' ,A+B)
    i+=1
```

역시나 틀렸다. 그래서 print문의 포맷이 잘못됐다는 것을 깨딷고 print문을 고치고자 했다.

```python
T=int(input())
for i in range(1, T+1):
    A, B = map(int, input().split())
    print('Case #',str(i),':' ,A ,'+' ,B ,'=' ,A+B)
```

우선Case와 #은 분리했어야 했고 str(i)를 썼다면 +를 써서 연결해야 했는데 그렇게 하지 못했다. 결국 찾아서 해결했다.
