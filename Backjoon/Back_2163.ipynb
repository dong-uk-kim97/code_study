정답

```python
# 입력
N, M = map(int,input().split())
# 출력
print(N*M-1)
```

오답

```python
# 입력
N,M = map(int,input().split())
# 출력
print(N*M-(N*M-1))
```

최소 쪼개기한 횟수를 구해야 하는데 1이 남으려면 어떻게 해야하는지에 집중하다보니 답을 이상하게 제출했다. 문제를 조금 더 깊이 이해해야겠다는 자세가 필요하다는 insight를 얻었다.

숏코딩

```python
print(eval(input().replace(' ','*'))-1)
```

eval 함수는 식을 입력하면 답을 출력하는 함수이다. 그렇기 때문에 공백을 곱하기로 계산한 값을 그대로 출력시켜서 -1시켜주는 것이다. 이 함수를 알고는 있었는데 실질적으로 써먹지를 못했다. 이런 방식도 생각을 해보는 것도 좋겠다.
