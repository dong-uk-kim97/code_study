# 최초 시도

```python
import math
T=int(input())
A,B = map(int,input().split())
print(A*B/math.gcd(A,B))
```

최초 시도에서 틀렸던 이유는 T갯수만큼 반복을 해야하는데 반복을 사용하지 않았다. 그렇기 때문에 반복문을 넣어서 2차 시도를 했다. 

# 2차 시도

```python
import math
T=int(input())
for _ in range(1,T+1):
    A,B = map(int,input().split())
    print(A*B/math.gcd(A,B))
```

for 반복문을 사용했는데도 틀렸다. 생각을 해보니 그냥 나누기를 사용하면 실수가 나온다는 것을 깜빡했다. 그래서 정수형으로 바꾸기 위해 몫만 나오는 나누기인 //를 사용하여 재시도했다.

# 정답

```python
import math
import sys
T=int(input())
for _ in range(1,T+1):
    A,B = map(int,sys.stdin.readline().split())
    print(A*B//math.gcd(A,B))
```

위 시도에서 정답이 됐다. 혹시나 시간이 초과될까봐 sys.stdin.readline()을 사용했다. 그리고 정수형으로 출력하기 위해 // 연산을 사용했다.

# 회고

math 라이브러리를 사용해서 최대공약수를 구하는 math.gcd()메소드를 활용하니 답을 코드를 쉽게 구할 수 있었다. 문제의 조건을 꼼꼼히 확인해야겠다는 생각을 했다. 정수형으로 출력하기 위해 //연산을 사용해야 하는 점 역시 숙지했다.
