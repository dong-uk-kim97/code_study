```python
n, x = map(int, input().split())
A = list(map(int,input().split()))
for i in A:
  if i<x:
    print(i, end= " ")
```

```python
n,x = map(int, input().split())
for i in range(1,n+1):
    if i<x:
        print(i)
```

빈 수열을 만들지 않았기 때문에 잘못됐다. 그리고 수열 안에 있는 숫자 중 x보다 작은 것을 골라야 하기 때문에 잘못됐다.
